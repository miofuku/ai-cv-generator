### 21211-03: Vorzeitige Lösungen von Ausbildungsverträgen, Geschlecht, Nationalität, Ausbildungsbereiche - Zeitreihe

```python
sheet_name =  '21211-03'
df = pd.read_excel(file_path, sheet_name=sheet_name, header=[3, 4])
df = df.rename(columns={'Unnamed: 0_level_1': 'Beruf', 'Unnamed: 1_level_1': 'Jahr'})
df[('Ausbildungsbereich / Jahr', 'Beruf')] = df[('Ausbildungsbereich / Jahr', 'Beruf')].fillna(method='ffill')
groups = df.groupby(('Ausbildungsbereich / Jahr', 'Beruf'))
```

#### 1. Premature Terminations of Training Contracts Over Time for All

```python
plt.figure(figsize=(10, 6))

for group_name, group_data in groups:
    plt.plot(group_data[('Ausbildungsbereich / Jahr','Jahr')], group_data[('Insgesamt','Insgesamt')], marker='o', label=group_name)
    
plt.title('Premature Terminations of Training Contracts Over Time for All')
plt.xlabel('Year')
plt.ylabel('Number')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Ratio of German in All Premature Terminations of Training Contracts

```python
plt.figure(figsize=(10, 6))

for group_name, group_data in groups:
    plt.plot(group_data[('Ausbildungsbereich / Jahr','Jahr')], group_data[('Insgesamt','Deutsche')] / group_data[('Insgesamt','Insgesamt')], marker='o', label=group_name)
    
plt.title('Ratio of German in All Premature Terminations of Training Contracts')
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 3. Ratio of Foreigners in All Premature Terminations of Training Contracts

```python
plt.figure(figsize=(10, 6))

for group_name, group_data in groups:
    plt.plot(group_data[('Ausbildungsbereich / Jahr','Jahr')], group_data[('Insgesamt','Ausländer')] / group_data[('Insgesamt','Insgesamt')], marker='o', label=group_name)
    
plt.title('Ratio of Foreigners in All Premature Terminations of Training Contracts')
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 4. Number of German and Foreigner in Premature Terminations of Training Contracts

```python
fig, ax = plt.subplots(figsize=(10,6))

bar_width = 0.35
index = np.arange(7)

years = range(2012, 2023)
foreigner = df[('Insgesamt','Ausländer')]
german = df[('Insgesamt','Deutsche')]
group = set(df[('Ausbildungsbereich / Jahr', 'Beruf')])

foreigner_value = []
german_value = []
for i in range(0, len(foreigner), 7):
    foreigner_value.append(foreigner[i:i+7])
    german_value.append(german[i:i+7])

for i, year in enumerate(years):
    ax.bar(index+bar_width, foreigner_value[i], bar_width, label=f'Foreigner ({year})')
    ax.bar(index, german_value[i], bar_width, label=f'German ({year})')

ax.set_xticks(index+bar_width/2)
ax.set_xticklabels(group)

ax.set_xlabel('Job')
ax.set_ylabel('Number')
ax.set_title('Number of German and Foreigner in Premature Terminations of Training Contracts')
ax.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
