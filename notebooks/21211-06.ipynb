### 21211-06: Fortbildungs-/Meisterprüfungen, Geschlecht, Prüfungserfolg, Ausbildungsbereiche - Zeitreihe
进修通过

```python
sheet_name =  '21211-06'
df = pd.read_excel(file_path, sheet_name=sheet_name, header=[3, 4])
df = df.rename(columns={'Unnamed: 0_level_1': 'Beruf', 'Unnamed: 1_level_1': 'Jahr'})
df[('Ausbildungsbereich / Jahr', 'Beruf')] = df[('Ausbildungsbereich / Jahr', 'Beruf')].fillna(method='ffill')
groups = df.groupby(('Ausbildungsbereich / Jahr', 'Beruf'))
print(df.columns)
```

#### 1.Continuing education passings over time

进修通过的总人数随时间变化


```python
plt.figure(figsize=(8, 6))

for group_name, group_data in groups:
    plt.plot(group_data[('Ausbildungsbereich / Jahr','Jahr')], group_data[('Insgesamt', 'Bestandene\nPrüfungen')], marker='o', label=group_name)

plt.title('Continuing education passings over time')
plt.xlabel('Year')
plt.ylabel('Number')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2.Continuing education pass rate over time

进修总通过率随时间变化

```python
plt.figure(figsize=(10, 6))

for group_name, group_data in groups:
    plt.plot(group_data[('Ausbildungsbereich / Jahr','Jahr')], group_data[('Insgesamt', 'Bestandene\nPrüfungen')] / group_data[('Insgesamt', 'Prüfungs-\nteilnahmen')] , marker='o', label=group_name)

plt.title('Continuing education pass rate over time')
plt.xlabel('Year')
plt.ylabel('rates')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
